```python
import pandas as pd

WEB_DEVLOPER_SKILLS = set(['django','html','css','nodejs','reactjs','nodejs','flask','laravel'])

MACHINE_LEARNING_SKILLS = set(['nosql','pandas','numpy','matpotlib','seabron','scipy'])


def skill_assign(df):

    designations = []
    for skills in df['Skills']:
        skills, designation = set(map(lambda x: x.lower(), eval(skills))), []
        ml_skills = skills.intersection(MACHINE_LEARNING_SKILLS)
        web_skills = skills.intersection(WEB_DEVLOPER_SKILLS)

        if len(ml_skills) >= 3:
            designation.append('Machine learning developer')
        elif len(web_skills) >= 3:
            designation.append('Web Developer')
        else:
            designation.append('na')
        designation = ', '.join(designation)
        designations.append(designation)

    df['Designation'] = designations
    df.to_excel('Designated.xlsx')
```

```python
filename = '/content/skill_data.xlsx'
skill_assign(pd.read_excel(filename,index_col=0))
```
